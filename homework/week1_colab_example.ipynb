### II. Methods

1. Generate 2 list of random integers. One in the range of 0-19 and the other in the range of 15-29. Each has 500 and 1000 realizations respectively.
2. Make histograms of these two lists of integers to examine the distributions.

```python
# Import the packages used in this example.

import random
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generate 500 random integers in the range 0-19.

random_numbers_list1 = []
for i in range(0, 500):
    random_numbers_list1.append(random.randint(0, 20))

# Generate 1000 random integers in the range 15-29.
random_numbers_list2 = []
for i in range(0, 1000):
    random_numbers_list2.append(random.randint(10, 30))
```

```python
# Create a dataframe for downstream analysis
df = pd.DataFrame(data={"randomly generated integers":random_numbers_list1+random_numbers_list2, "batch": [1]*500 + [2]* 1000} )
display(df.head()) # Display the first few rows
```

### III. Results
Visualize the randomly generated integers.

```python
# Plot the histogram of these randomly generated integers.
plt.figure(dpi=150)
sns.histplot(data=df, x="randomly generated integers", hue="batch", bins=31)
plt.show()
```

### IV. Conclusion and Discussion
We generated 2 sets of random integers, one from the range 0 to 19 and one from the range 15 to 29. These random integers are mostly uniformly distributed. However, due to stochasticity, the number of times each integer appears are not exactly the same.
